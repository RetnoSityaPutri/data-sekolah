```python
import requests
import json

# URL API untuk mengambil data sekolah
url = "https://api-sekolah-indonesia.vercel.app/sekolah?page=1&perPage=1000"

# Permintaan GET untuk mengambil data dari API
response = requests.get(url)

# Mengecek apakah permintaan berhasil
if response.status_code == 200:
    # Parsing data JSON dari respons
    data = response.json()

    # Menyimpan data dalam format JSON
    with open('data_sekolah.json', 'w') as json_file:
        json.dump(data, json_file, indent=4, ensure_ascii=False)

    print("Data berhasil diambil dan disimpan dalam file data_sekolah.json.")
else:
    print(f"Gagal mengambil data. Status code: {response.status_code}")
```

```python
import json

# Membaca file JSON
with open('/content/data_sekolah.json') as file:
    data = json.load(file)

```

```python
import pandas as pd

# Mengakses data utama yang relevan
data_sekolah = data['dataSekolah']  # ganti 'data' sesuai dengan kunci yang benar

# Konversi ke DataFrame
df = pd.DataFrame(data_sekolah)

# Menampilkan DataFrame
display(df)
```
